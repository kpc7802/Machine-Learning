```python
#import data

import tensorflow as tf
fmnist = tf.keras.datasets.fashion_mnist
```

```python
(x_training_data, y_training_data), (x_validation_data, y_validation_data) = fmnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40,40))

num_images = 20
for i in range(num_images):
    row = x_training_data[i]
    label = y_training_data[i]

    image = row.reshape(28,28)
    plt.subplot(1, num_images, i+1)
    plt.title(label, fontdict={'fontsize': 30})
    plt.axis('off')
    plt.imshow(image, cmap='gray')
```

```python
#x_training_data.shape

x_training_data = x_training_data.reshape(60000, 784)
x_validation_data = x_validation_data.reshape(10000, 784)

#x_training_data.max()

x_train_norm = x_training_data / 255
x_valid_norm = x_validation_data / 255

#x_train.max()
```

```python
#imports for future steps
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
num_categories = 10

y_training_data = keras.utils.to_categorical(y_training_data, num_categories)
y_validation_data = keras.utils.to_categorical(y_validation_data, num_categories)
```

```python
model = Sequential()
model.add(Dense(units = 500, activation = 'relu'))
model.add(Dense(units = 500, activation = 'relu'))
model.add(Dense(units = 10, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
epoch = model.fit(x_train_norm, y_training_data, epochs=15, verbose=1, validation_data=(x_validation_data, y_validation_data))
```
